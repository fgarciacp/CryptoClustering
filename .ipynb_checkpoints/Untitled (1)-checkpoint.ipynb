```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import hvplot.pandas  
from sklearn.decomposition import PCA
```

```python
# Cargar el archivo CSV en un DataFrame
df_crypto = pd.read_csv('C:/Users/Usuario/Downloads/Starter_Code (16)/Starter_Code/Resources/crypto_market_data.csv')
```

```python
# Obtener las estad√≠sticas de resumen
summary_statistics = df_crypto.describe()
summary_statistics
```

```python
print(df_crypto.columns)
```

```python

# Selecciona las columnas que deseas graficar
columns_to_plot = ['price_change_percentage_24h', 'price_change_percentage_7d',
                   'price_change_percentage_14d', 'price_change_percentage_30d',
                   'price_change_percentage_60d', 'price_change_percentage_200d',
                   'price_change_percentage_1y']

# Crea un gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
for column in columns_to_plot:
    plt.plot(df_crypto['coin_id'], df_crypto[column], label=column)

plt.title('Variaci√≥n del Precio de las Criptomonedas')
plt.xlabel('ID de la Criptomoneda')
plt.ylabel('Cambio de Precio (%)')
plt.legend()
plt.show()
```

```python
df_crypto.set_index('coin_id', inplace=True)
```

```python
# Seleccionar las columnas que deseas escalar
columns_to_scale = ['price_change_percentage_24h', 'price_change_percentage_7d',
                    'price_change_percentage_14d', 'price_change_percentage_30d',
                    'price_change_percentage_60d', 'price_change_percentage_200d',
                    'price_change_percentage_1y']
```

```python
# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_crypto[columns_to_scale])
```

```python
# Crear un nuevo DataFrame con los datos escalados
df_scaled = pd.DataFrame(scaled_data, index=df_crypto.index, columns=columns_to_scale)

# Mostrar las primeras cinco filas del DataFrame escalado
df_scaled.head()
```

```python
# Crear un gr√°fico de l√≠neas con los datos escalados
plt.figure(figsize=(10, 6))
for column in columns_to_scale:
    plt.plot(df_scaled.index, df_scaled[column], label=column)

plt.title('Variaci√≥n del Precio Escalado de las Criptomonedas')
plt.xlabel('ID de la Criptomoneda')
plt.ylabel('Cambio de Precio Escalado')
plt.legend()
plt.show()
```

```python
# Calcular la inercia para cada valor de k
k_values = range(1, 11) 
inertia = []  
```

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
# Crear un diccionario con los datos para trazar la curva del codo
elbow_data = {'k': k_values, 'inertia': inertia}
```

```python
# Dibujar un gr√°fico lineal con todos los valores de inercia
plt.figure(figsize=(10, 6))
plt.plot(elbow_data['k'], elbow_data['inertia'], marker='o')
plt.title('Curva del Codo')
plt.xlabel('N√∫mero de Cl√∫steres (k)')
plt.ylabel('Inercia')
plt.xticks(elbow_data['k'])  
plt.grid()
plt.show()
```

```python
# ¬øCu√°l es el mejor valor para k? =  4
```

```python
# Inicializar el modelo K-means con el mejor valor para k
best_k = 4  # Cambia esto al mejor valor de k que encontraste
kmeans = KMeans(n_clusters=best_k, n_init=10, random_state=42)
```

```python
# Ajustar el modelo K-means utilizando el DataFrame escalado
kmeans.fit(scaled_data)
```

```python
# Predecir los cl√∫steres para agrupar las criptomonedas
clusters = kmeans.predict(scaled_data)
```

```python
df_crypto.reset_index(inplace=True)
```

```python
# Crear una copia del DataFrame escalado y agregar una nueva columna con los cl√∫steres previstos
df_scaled_with_clusters = pd.DataFrame(scaled_data, columns=columns_to_scale)
df_scaled_with_clusters['coin_id'] = df_crypto['coin_id'].values  
df_scaled_with_clusters['cluster'] = clusters  
```

```python
# Crear un gr√°fico de dispersi√≥n utilizando hvPlot
scatter_plot = df_scaled_with_clusters.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='cluster',  
    hover_cols=['coin_id'], 
    title='Agrupamiento de Criptomonedas',
    xlabel='Cambio de Precio en 24h (%)',
    ylabel='Cambio de Precio en 7d (%)',
    height=400,
    width=600
)
```

```python
# Mostrar el gr√°fico
scatter_plot.opts(tools=['hover'])  
scatter_plot
```

```python
# Aplicar PCA
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
```

```python
# Calcular la varianza explicada
explained_variance = pca.explained_variance_ratio_
total_explained_variance = explained_variance.sum()
print("Varianza total explicada por los tres componentes principales:", total_explained_variance)
```

```python
df_pca = pd.DataFrame(data=pca_data, index=df_crypto.index, columns=['PC1', 'PC2', 'PC3'])
```

```python
df_pca
```

```python
# Crear una lista con el n√∫mero de valores k del 1 al 12
k_values = range(1, 12)  # k de 1 a 11
inertia = []  
```

```python
# Calcular la inercia para cada valor de k
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(df_pca)  
    inertia.append(kmeans.inertia_)  
```

```python
#   diccionario para trazar la curva del codo
elbow_data = {
    'k': k_values,
    'inertia': inertia
}
```

```python
#  Dibujar un gr√°fico lineal
plt.figure(figsize=(10, 6))
plt.plot(elbow_data['k'], elbow_data['inertia'], marker='o')
plt.title('M√©todo del Codo')
plt.xlabel('N√∫mero de Cl√∫steres (k)')
plt.ylabel('Inercia')
plt.xticks(k_values)
plt.grid()
plt.show()
```

```python
# Mostrar el gr√°fico
scatter_plot.opts(tools=['hover'])
scatter_plot
```
